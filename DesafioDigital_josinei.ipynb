```python
import pandas as pd
import plotly.express as px
import os
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

**Base de Dados**

```python
df = pd.read_excel('basepython.xlsx')
df
```

```python
df.info()
```

**Algumas limpezas e ajustes**

```python
df = df.drop(['Cod_vendedor','Vendedor'], axis=1)
df = df.rename(columns={'Cod_vendedorr': 'Vendedor'})
df = df.rename(columns={'Faturamento liquido': 'lucros liquidos'})
df = df.replace(['Ar_condicionado','Cadeira%Gamer', 'Fone d Ouvido', 'IPHOne', 'iPHONE','SAMSUNG', 'samsungsamsung','XBOX SERIESSSS', 'Amazonas Shoping'], ['Ar condicionado', 'Cadeira Gamer', 'Fone de Ouvido','Iphone', 'Iphone', 'Samsung', 'Samsung' ,'Xbox series s','Amazonas Shopping'])
```

```python
df['Centro'] = df['Centro'].astype(int)
df['Valor unitário'] = df['Valor unitário'].astype(int)
df['Qtd'] = df['Qtd'].astype(int)
df['Valor Venda'] = df['Valor Venda'].fillna(0).astype(int)
df['% Venda Total'] = df['% Venda Total'].fillna(0).astype(int)
```

```python
df
```

**QUAL IMPOSTO A SE PAGO POR UNIDADE E QUAL IMPOSTO TOTAL?**

```python
df1 = df.groupby('Unidade.1')['Valor Imposto'].sum().reset_index()
fig1 = go.Figure(data=[go.Bar(x=df1['Unidade.1'], y=df1['Valor Imposto'],marker_color=['blue', 'green', 'red', 'orange', 'purple','pink','green'] )])
fig1.show()
```

**Qual o produto que mais vende?**

```python
df2 = df.groupby(['Produto'])['Qtd'].sum().reset_index()
fig2 = go.Figure(go.Bar(
    x=df2['Qtd'],
    y=df2['Produto'],
    orientation='h',
    text=df2['Qtd'],
    textposition='auto',
    marker_color='blue'
))
fig2.update_layout(
    title='Quantidade Total de Produtos Vendidos',
    xaxis_title='Quantidade Total',
    yaxis_title='Produto',
    bargap=0.2,
    bargroupgap=0.1
)
fig2.show()
```
